```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
options = Options()
options.add_argument("--disable-gpu")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service = service, options = options)

url = "https://www.flipkart.com/search?q=apple%20iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
driver.get(url)
```

```python
product_title = []
product_price = []
product_discount = []
product_ratings = []
product_reviews = []
```

```python
num_pages = 97
```

```python
for item in range(num_pages):
    time.sleep(2)

    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")

    product = soup.find_all("div", class_ = "yKfJKb row")

    for i in product:
        name = i.find("div", class_ = "KzDlHZ")
        if name:
            product_title.append(name.get_text(strip=True))
        else:
            product_title.append(None)

        price = i.find("div", class_ = "Nx9bqj _4b5DiR")
        if price:
            product_price.append(price.get_text(strip=True))
        else:
            product_price.append(None)
            
        discount = i.find("div", class_ = "UkUFwK")
        if discount:
            product_discount.append(discount.get_text(strip=True))
        else:
            product_discount.append(None)

        ratings = i.find("span", class_ = "Y1HWO0")
        if ratings:
            product_ratings.append(ratings.get_text(strip=True))
        else:
            product_ratings.append(None)

        reviews = i.find("span", class_ = "Wphh3N")
        if reviews:
            product_reviews.append(reviews.get_text(strip=True))
        else:
            product_reviews.append(None)

    try: # for next page button xpath
        next_button = driver.find_element('xpath', '//a[@class="_9QVEpD"][span[text()="Next"]]')
        next_button.click()
    except Exception as e:
        print(f"Error finding or clicking in Next button.")
        break

driver.quit()
```

```python
data = {"Product" : product_title, "Price" : product_price,
        "Discount" : product_discount, "Ratings" : product_ratings, "Reviews" : product_reviews}
df = pd.DataFrame(data)
df
```
